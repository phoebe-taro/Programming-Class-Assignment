###ex32

####a

```python
def fib(n):
  n0 = 1
  n1 = 1
  for i in range(1,n+1):
    if i > 2 :
      n2 = n0+n1
      n0,n1 = n1,n2
      print(n2)
    else :
      print(1)

fib(8)
```

####b

```python
def fib_gen(n):
  n0 = 1
  n1 = 1
  for i in range(1,n+1):
    if i < 3:
      yield 1
    else:
      n2 = n0+n1
      n0,n1 = n1,n2
      yield n2

text = fib_gen(5)
print(next(text))
print(next(text))
print(next(text))
print(next(text))
print(next(text))
```

###ex33

```python
def gen(n):
  for i in range(0,n+1):
    if i % 3 == 0 and i % 5 != 0:
      yield i

text = gen(15)
print(next(text))
print(next(text))
print(next(text))
print(next(text))
```

###ex34

```python
def primes(n):
  for i in range(2,n+1):
    for j in range(2,i):
      if i % j == 0:
        break
    else :
      yield i

text = primes(10)
print(next(text))
print(next(text))
print(next(text))
print(next(text))
```

###ex35

```python
def primes():
  num = 2
  while True:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
      if num % i == 0:
        is_prime = False
        break
    if is_prime:
      yield num
    num += 1

g = primes()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

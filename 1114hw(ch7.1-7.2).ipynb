```python
class Sphere:
  def __init__(self,r):
    self.rad = r
  def volume(self):
    return 4/3*3.14*self.rad**2
  def surface_area(self):
    return 4*3.14*self.rad**2
  def __repr__(self):
    return f'Sphere object, rad = {self.rad} '
  def __str__(self):
    return f'Sphere object , rad = {self.rad} , volume = {Sphere.volume(self):.2f} , surface_area = {Sphere.surface_area(self):.2f}'
```

####a

```python
c1 = Sphere(5)
c1
```

####b

```python
print(c1)
```

###ex6

```python
import math
class Calculator :
  def __init__(self,a,b):
    self.n1 = a
    self.n2 = b
  def add(self):
    return self.n1 + self.n2
  def gcd(self):
    return math.gcd(self.n1,self.n2)
  def lcm(self):
    return self.n1 * self.n2 // Calculator.gcd(self)
  def power(self):
    return self.n1 ** self.n2
```

####a

```python
c0 = Calculator(2,10)
c0.add()
```

####b

```python
c0 = Calculator(2,10)
c0.gcd()
```

####c

```python
c0 = Calculator(2,10)
c0.lcm()
```

####d

```python
c0 = Calculator(2,10)
c0.power()
```

```python
def factors(n):
  return [i for i in range(1,n+1) if n % i == 0]
def primes(n):
  prime = []
  for i in range(1,n):
    for j in range(2,i):
      if i % j == 0 :
        break
    else:
      prime.append(i)
  return prime
```

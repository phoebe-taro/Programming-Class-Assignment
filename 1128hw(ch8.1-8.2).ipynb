```python
with open ('factors.txt','w') as f:
 factors = [ str(i) for i in range(1,101) if 100 % i == 0]
 f.writelines(','.join(factors))
```

####b

```python
with open ('factors.txt','r') as f:
  sum(eval(f.readline()))
```

####c

```python
with open ('factors.bin','wb') as f:
  factor = [ i for i in range(1,101) if 100 % i == 0]
  f.write(bytes(factor))
  print(bytes(factor))
```

####d

```python
with open ('factors.bin','rb') as f:
  print(sum(list(f.read())))
```

###ex3

```python
with open ('mat.txt','w') as f:
  f.writelines(['23 45 23 65\n',
               '44 56 88 21\n',
               '50 67 89 12\n'])
with open ('mat.txt','r') as f:
  txt = f.readlines()
  num = []
  for i in txt:
    numbers = i.strip().split()
    for j in numbers:
      num.append(int(j))
  print(f'最大值:{max(num)}')
  print(f'最小值:{min(num)}')
```

###ex5

####a

```python
lst = []
for i in range(2,1001):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    lst.append(i)
print(lst)
```

####b

```python
import pickle
with open('primes.bin','wb') as f :
  f.write(pickle.dumps(lst))
```

####c

```python
with open('primes.bin','rb') as f :
  prime = pickle.loads(f.readline())
print(prime)
```

###ex7

####a

```python
import csv
from io import StringIO
csv_data='75,87,86,19,69\n78,65,12,77,90'
f0=StringIO(csv_data)
data = [int(i) for j in csv.reader(f0,delimiter=',') for i in j]
print(f'最大值:{max(data)}')
```

####b

```python
import pickle
from io import BytesIO
data = [{3,5}, 'Python', (64,1024)]
barr = pickle.dumps(data)
f1 = BytesIO(barr)
print(pickle.loads(f1.read()))
```

###ex9

```python
while(True):
  a = input('請輸入第一個數字a:')
  if a in 'Qq':
    print('程式結束')
    break
  b = input('請輸入第二個數字b:')
  if b in 'Qq':
    print('程式結束')
    break
  else :
    try:
      if  int(a) < int(b) :
        print('a小於b')
      else:
        print('a大於b')
      print('判別完成')
    except:
      print('無法判別大小，請重新輸入')
```

###ex10

```python
n = input('請輸入一個數字:')
try:
  if int(n) < 0:
    print('不能輸入負數')
  else:
    print('輸入正確')
    print(int(n)**0.5)
except ValueError:
  print('必須輸入數字')
```

```python
def to_byteArray(lst):
  try:
    return bytes(lst)
  except:
    print('無法轉換')
```

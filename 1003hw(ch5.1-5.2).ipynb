###ex11

```python
triangle =eval(input('請讀入一個序對 (a, b, c)，代表三個邊長:'))
a = triangle[0]
b = triangle[1]
c = triangle[2]
if a + b > c and a + c > b and b + c > a :
  if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2 :
    print(f'{triangle}為直角三角形')
  else :
    print(f'{triangle}不是直角三角形')
else :
  print(f'{triangle}無法形成三角形')
```

###ex12

```python
price = int(input('請輸入購買金額:'))
coin  = 100 - price
if coin > 0 :
  fifty = coin  // 50
if coin % 50 > 0 :
  ten = coin % 50 // 10
if coin % 50 % 10 > 0 :
  five = coin % 50 % 10 // 5
print(f'50元硬幣{fifty}枚，10元硬幣{ten}枚，5元硬幣{five}枚，1元硬幣{coin%50%10%5}枚')
```

```python
a = int(input('請輸入第一個數:'))
b = int(input('請輸入第二個數:'))
if a > b:
  a, b = b, a
for i in range(1, a + 1):
  if a % i == 0 and b % i == 0:
    gcd = i
print(f'最大公因數為{gcd}')
```

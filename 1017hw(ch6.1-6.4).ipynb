```python
def isprime(x):
  i = 2
  j = True
  while i < x:
    if x % i == 0:
      j = False
      break
    i += 1
  return j
```

```python
a = int(input("請輸入一個數字:"))
isprime(a)
```

###ex4

```python
def primes(x):
  a = []
  for i in range(2,x+1):
    for j in range(2,i):
      if i % j == 0:
        break
    else:
      a.append(i)
  return a
```

```python
a = int(input("請輸入一個數字:"))
primes(a)
```

```python
def prime_factors(x):
  a = []
  j = 2
  for i in range(2,x+1):
      if x % i == 0:
        for j in range(2,i):
          if i % j == 0 :
            break
        else:
          a.append(i)
  return a
```

```python
a = int(input("請輸入一個數字:"))
prime_factors(a)
```

###ex6

```python
def factor(x):
  a = []
  i = 2
  while i*i <= x:
    c = 0
    while x % i == 0:
      c += 1
      x //= i
    if c > 0:
      a.append((i,c))
    i += 1
  if x > 1:
    a.append((x,1))
  return a
```

```python
a = int(input("請輸入一個數字:"))
factor(a)
```

###ex7

```python
def bin2dec(bs):
  l = len(bs) - 1
  a = 0
  if l+1 == 0 :
    print('None')
  else:
    for i in bs:
      a += int(i) * 2**l
      l -= 1
    print(a)
```

```python
a = input("請輸入一個二進位數字:")
bin2dec(a)
```

```python
def to_numbes(word):
  n = {2:('A','B','C'),
         3:('D','E','F'),
         4:('G','H','I'),
         5:('J','K','L'),
         6:('M','N','O'),
         7:('P','Q','R','S'),
         8:('T','U','V'),
         9:('W','X','Y','Z')}
  for i in word:
    for j in range(2,10):
        if i in n.get(j):
          print(j,end='')
```

```python
a = input("請輸入大寫字串:")
to_numbes(a)
```

###ex16  

```python
def euler_pi(n=10):
  pi = 2
  primes = []
  num = 3
  while len(primes) < n:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
      if num % i == 0:
        is_prime = False
        break
    if is_prime:
      primes.append(num)
      numerator = num
      if (numerator+1)%4 == 0:
        denominator = numerator-1
      else:
        denominator = numerator+1
    pi *= (numerator/denominator)
    num += 2
  print('計算到',numerator,'/',denominator)
  return pi
```

```python
euler_pi(5)
```

###ex17

```python
def make_dict(keys,values=0):
  if values == 0:
    values = [0]*len(keys)
  d = {}
  for i,j in zip(keys,values):
    d.update({i:j})
  return d
```

```python
make_dict([0,1,2], [32,43,55]),make_dict([0, 1, 2])
```

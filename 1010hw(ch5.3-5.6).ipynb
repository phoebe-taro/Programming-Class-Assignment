```python
num = int(input('請輸入一個整數:'))
digits = 0
while num>= 10:
    num //= 10
    digits += 1
print(f'{digits+1}個位數的整數')
```

###ex21

```python
numstr = input('請輸入一個帶有千分位符號的整數:')
new = ''
for char in numstr:
  if char != ',':
    new += char
num = int(new)
print(num * 2)
```

###ex22

```python
num = int(input('請輸入一個整數:'))
n = 1
factor = ''
digits = 0
while n <= num:
  if num % n == 0:
    factor += str(n) + ','
    digits += 1
  n += 1
print(f'{num}的因數有{digits}個，分別為{factor.strip(",")}')
```

###ex23

```python
num = int(input('請輸入一個整數:'))
n = 2
prime = True
while n < num :
  if num % n == 0:
    prime = False
    break
  n += 1

if prime == True:
  print(f'{num}是質數')
else:
  print(f'{num}不是質數')
```

###ex31

```python
lower_limit=0
upper_limit=255
tandem = eval(input('請輸入一串列:'))
for n in tandem:
  if not(lower_limit <= n <= upper_limit):
    print('最少有一個數不在 “[0, 255] 之內')
    break
else:
  print('輸入的數全在範圍之內')
```

```python
x = np.linspace(0,np.pi,32)
y = np.exp(-0.5*x)*np.cos(3*x)
fig,ax = plt.subplots()
ax.plot(x,y,'r--o',markersize=8)
ax.fill_between(x,y,alpha=0.5,color='yellow')
ax.set_title('My plot',fontsize=18)
ax.grid()
plt.show()
```

###ex18

####a

```python
x = np.linspace(0,2*np.pi)
y = np.linspace(0,2*np.pi)
XX,YY = np.meshgrid(x,y)
ZZ = np.sin(XX)*np.sin(YY)
colors = ['red','blue','green']
line = [-0.7,-0.3,0,0.3,0.7]
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,3,1)
cs =ax1.contour(XX,YY,ZZ,line)
ax1.clabel(cs,inline=1,fontsize=8)

ax2 = fig.add_subplot(1,3,2,projection='3d')
ax2.contour3D(XX,YY,ZZ,line)

ax3 = fig.add_subplot(1,3,3,projection='3d')
ax3.contour3D(XX,YY,ZZ,line)
ax3.plot_surface(XX,YY,ZZ,cmap='jet',alpha=0.7)
plt.show()
```

####b

```python
x = np.linspace(-np.pi,np.pi)
y = np.linspace(-np.pi,np.pi)
XX,YY = np.meshgrid(x,y)
ZZ = 1/np.sqrt(XX**2+YY**2+1)
colors = ['red','blue','green']
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,3,1)
cs = ax1.contour(XX,YY,ZZ,5)
ax1.clabel(cs,inline=1,fontsize=8)

ax2 = fig.add_subplot(1,3,2,projection='3d')
ax2.contour3D(XX,YY,ZZ,5)

ax3 = fig.add_subplot(1,3,3,projection='3d')
ax3.contour3D(XX,YY,ZZ,5)
ax3.plot_surface(XX,YY,ZZ,cmap='jet',alpha=0.7)
plt.show()
```

```python
x = np.linspace(-4,4)
y = np.linspace(-4,4)
XX,YY = np.meshgrid(x,y)
ZZ = YY/(XX**2+YY**2+1)**(4/5)
colors = ['red','blue','green']
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,3,1)
cs = ax1.contour(XX,YY,ZZ,6)
ax1.clabel(cs,inline=1,fontsize=8)

ax2 = fig.add_subplot(1,3,2,projection='3d')
ax2.contour3D(XX,YY,ZZ,6)

ax3 = fig.add_subplot(1,3,3,projection='3d')
ax3.contour3D(XX,YY,ZZ,6)
ax3.plot_surface(XX,YY,ZZ,cmap='jet',alpha=0.7)
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

###ex22

```python
x = np.linspace(0,np.pi,240)
y = np.tan(np.sin(x))-np.sin(np.tan(x))
fig,ax = plt.subplots()
dot,=ax.plot([],[],'ro',markersize=10)
plt.close()
def init():
  ax.plot(x,y)
def animate(i):
  dot.set_data(x[i],y[i])
ani = FuncAnimation(fig=fig,func=animate,frames=240,init_func=init,interval=50)
HTML(ani.to_html5_video())
```

###ex23

```python
t=np.linspace(0,2*np.pi,256)
r=np.sin(6*t)
fig,ax = plt.subplots(subplot_kw={'projection':'polar'})
dot,=ax.plot([],[],'ro',markersize=10)
line, = ax.plot([],[])
plt.close()
def animate(i):
  dot.set_data(t[i],r[i])
  line.set_data(t[:i],r[:i])
ani = FuncAnimation(fig=fig,func=animate,frames=256,interval=50)
HTML(ani.to_html5_video())
```

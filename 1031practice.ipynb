```python
string = input('請輸入一個字串:')
s1 = ''
d1 = {}
for i in string:
  if i.isalpha():
    s1 = s1 + i
  else :
    continue
s1 = s1.lower()
for j in s1:
  c = s1.count(j)
  d1[j] = c
print(d1)
```

```python
import random
ans = random.randint(1,10)
count = 0
player = 0
while player != ans:
  player = int(input('你的猜測:'))
  if player < ans :
    print('你的猜測太小了，再試一次')
    count += 1
  elif player > ans :
    print('你的猜測太大了，再試一次')
    count += 1
  else:
    print(f'恭喜你！猜對了,正確數字是 {ans} 你總共猜了 {count+1} 次')
    break
```

```python
def harmonic_sum(n):
  if n > 1:
    return 1/n + harmonic_sum(n-1)
  else :
    return 1
```

```python
harmonic_sum(5)
```

```python
def natural_numbers():
  a = 1
  yield a
  while True:
    a += 1
    yield a
```

```python
gen = natural_numbers()
print(next(gen))
print(next(gen))
```

```python
def find_first_digit(s):
  for i in s:
    if  i.isdigit():
      return i
      break
  else:
    return 'None'
```

```python
print(find_first_digit("abc123"))
print(find_first_digit("hello456"))
print(find_first_digit("no_digits_here"))
```

####15

```python
def key_of_max_value(d):
  if d == {}:
    return 'None'
  else:
    maxvalue = max(d.values())
    for key,value in d.items():
      if value == maxvalue:
        return key
```

```python
print(key_of_max_value({'a': 1, 'b': 5, 'c': 3}))
print(key_of_max_value({'x': 10, 'y': 20, 'z': 5}))
print(key_of_max_value({}))
```

```python
def find_target(nums,  target):
  for i in nums:
    if i == target:
      return "Target found!"
      break
  else:
    return "Target not found!"
```

```python
print(find_target([1, 2, 3, 4, 5], 3))
print(find_target([10, 20, 30, 40], 25))
```

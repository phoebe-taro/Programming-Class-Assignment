```python
a = np.array([[0,1,2,3,4],
              [2,3,4,0,1],
              [4,0,1,2,3],
              [1,2,3,4,0],
              [3,4,0,1,2]])
```

####a

```python
r,c = 2,2
row,col = np.where(a==4)
dist = np.argmin((row-r)**2+(col-c)**2)
row[dist],col[dist]
```

####b

```python
r,c = 4,3
row,col = np.where(a==0)
dist = np.argmin((row-r)**2+(col-c)**2)
row[dist],col[dist]
```

####c

```python
r,c = 4,0
row,col = np.where(a==2)
dist = np.argmin((row-r)**2+(col-c)**2)
row[dist],col[dist]
```

####d

```python
row,col = np.where(a==4)
dist = np.zeros((len(row), len(row)))
for i in range(len(row)):
    for j in range(i + 1, len(row)):
        dist[i, j] = np.sqrt((row[i] - row[j])**2 + (col[i] - col[j])**2)
        dist[j, i] = dist[i, j]
max_dist_index = np.unravel_index(np.argmax(dist), dist.shape)
(row[max_dist_index[0]],col[max_dist_index[0]]),(row [max_dist_index[1]],col[max_dist_index[1]])
```

###ex6

```python
def prime_list(n):
  prime = []
  for i in range(1,n):
    for j in range(2,i):
      if i % j == 0 :
        break
    else:
      prime.append(i)
  return prime
```

```python
prime_list(30)
```

###ex7

```python
def find_pi(n):
    k=np.arange(0,n)
    return (1/16**k*(4/(8*k+1)-2/(8*k+4)-1/(8*k+5)-1/(8*k+6))).sum()
```

```python
find_pi(2)
```

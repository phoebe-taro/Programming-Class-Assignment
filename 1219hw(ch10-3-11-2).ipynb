###ex10

```python
def primes(n):
  is_prime = []
  for i in range(2,n):
    for j in range(2,i):
      if i % j == 0 :
        break
    else:
      is_prime.append(i)
  return is_prime

primes(50)
```

```python
n = np.linspace(2, 100, dtype=int)
y = [len(primes(i)) for i in n]
plt.plot(n, y,'.')
plt.show()
```

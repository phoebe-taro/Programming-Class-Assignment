```python
class Employee:
  payRate=[1, 1.2, 1.5]
  def __init__(self,base):
    self.baseSalary=base
  def salary(self,hr,bonus):
    return self.baseSalary*hr*__class__.payRate[bonus]
  @classmethod
  def set_payRate(cls,new_payRate):
    cls.payRate = new_payRate
  @staticmethod
  def estimate(bs,hr,rate):
    return bs*hr*rate
```

```python
tom=Employee(160)
print(tom.salary(10,1))
Employee.set_payRate([1,1.3,1.5])
print(tom.salary(10,1))
print(Employee.estimate(160,10,1.25))
```

```python
from os import lstat
class Factor:
  factor_list=[2,3,6,8]
  def __init__(self,num):
    print('initial factor_list:',Factor.factor_list)
    self.num=num
  def find_factors(self):
    factors = [i for i in Factor.factor_list if self.num % i == 0 ]
    return  factors
  @classmethod
  def add_factors(cls,lst):
    for i in lst:
      if i in Factor.factor_list:
        pass
      else:
        Factor.factor_list.append(i)
  @classmethod
  def remove_factors(cls,lst):
    for i in lst:
      if i in Factor.factor_list:
        Factor.factor_list.remove(i)
  @classmethod
  def show_factor_list(cls):
    return Factor.factor_list
  @staticmethod
  def isfactor(num,n):
    return num % n == 0
```

```python
f0=Factor(60)
print(f0.find_factors())
Factor.add_factors([3,9])
print(Factor.show_factor_list())
Factor.remove_factors([2,4])
print(Factor.show_factor_list())
print(f0.find_factors() )
Factor.isfactor(45,8)
```

###ex9

```python
class Car:
  def __init__(self,color):
    self.color=color
  def show(self):
    print(f'color={self.color}')
class Truck(Car):
  def __init__(self,dr,ow,co):
    self.doors=dr
    self.owner=ow
    super().__init__(co)
  def show(self):
    print(f'doors={self.doors},owner={self.owner},color={self.color}')
```

```python
Car('red').show()
Truck(2,'Tom','blue').show()
```

###ex10

```python
class Person:
  def __init__(self, na):
    self.name = na
  def print_name(self):
    print("name=",self.name)
class Student(Person):
  def __init__(self,name,gender):
    super().__init__(name)
    self.gender = gender
  def print_info(self):
    print("name=",self.name,"gender=",self.gender)
```

```python
s0=Student('Mary','F')
s0.print_name()
s0.print_info()
```

```python
#更改前
class Triangle:
  def __init__(self,base,height):
    self.__base=base
    self.__height=height
  def area(self):
    return 0.5*self.__base*self.__height
```

```python
t0=Triangle(10,5)
t0.area()
```

```python
t0.__base=40
t0.area()
#因為__base是私有屬性無法這樣被設定
```

```python
#更改後
class Triangle:
  def __init__(self,base,height):
    self.set_base(base)
    self.set_height(height)
  def area(self):
    return 0.5*self.__base*self.__height
  def set_base(self,base):
    self.__base=base
  def set_height(self,height):
    self.__height=height
```

```python
t0=Triangle(10,5)
t0.set_base(40)
t0.area()
```

```python
class Year:
  def __init__(self,y):
    self.__year=y
  def isleap(self):
    return  self.__year % 400 == 0 or (self.__year%4 == 0 and self.__year % 100 != 0)
  @property
  def year(self):
    return self.__year
  @year.setter
  def  year(self,y):
    self.__year = y
```

```python
y0=Year(2022)
print(y0.year)
print(y0.isleap())
y0.year=2020
print(y0.year)
print(y0.isleap())
```
